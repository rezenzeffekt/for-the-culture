```python
# imports
import pandas as pd
# reading the csv into a dataframe, requires data/Top100_Germany.csv & data/Top100_Mexico.csv
de = pd.read_csv('data/Top100_Germany.csv', header=0, index_col=False)
mx = pd.read_csv('data/Top100_Mexico.csv', header=0, index_col=False)
```

```python
display(de.tail(3), mx.tail(3))
```

```python
de1 = de.query('rank == 1').copy()
mx1 = mx.query('rank == 1').copy()
```

```python
display(de.shape, de1.shape, mx.shape, mx1.shape)
```

```python
de['dtime'] = pd.to_datetime(de['date'], format='%Y-%m-%d')
de1['dtime'] = pd.to_datetime(de1['date'], format='%Y-%m-%d')
mx['dtime'] = pd.to_datetime(mx['date'], format='%Y-%m-%d')
mx1['dtime'] = pd.to_datetime(mx1['date'], format='%Y-%m-%d')
```

```python
de.groupby(de['dtime'].dt.to_period('Q'))['streams'].sum().plot(kind='bar', figsize=(9,9))
```

```python
mx.groupby(mx['dtime'].dt.to_period('Q'))['streams'].sum().plot(kind='bar', figsize=(9,9))
```

```python
de1.groupby(de1['dtime'].dt.to_period('Q'))['streams'].sum().plot(kind='bar', figsize=(9,9))
```

```python
mx1.groupby(mx1['dtime'].dt.to_period('Q'))['streams'].sum().plot(kind='bar', figsize=(9,9))
```
