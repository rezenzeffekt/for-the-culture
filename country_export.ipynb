```python
# exporting to csv
from os.path import expanduser
from pathlib import Path

def get_name(variable):
    for name in globals():
        if id(globals()[name]) == id(variable):
            return name
    for name in locals():
        if id(locals()[name]) == id(variable):
            return name
    return None

def save(df):
    """Saves a dataframe to the Downloads folder"""
    home = Path(expanduser("~"))
    path = home / "Downloads"
    df_name = get_name(df) + ".csv"
    file_name = path / df_name
    df.to_csv(file_name, index=False)
    print(file_name) 
```

```python
save(Top100_UnitedStates)
save(Top100_SouthAfrica)
save(Top100_Sweden)
save(Top100_Australia)
save(Top100_SouthKorea)
save(Top100_Japan)
save(Top100_Mexico)
```
